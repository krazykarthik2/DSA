```python
import random
import time

# Function to print the Sudoku grid
def print_grid(grid):
    for row in grid:
        print(" ".join(str(num) if num != 0 else '.' for num in row))
    print("\n")

# Function to check if a number can be placed in a cell
def is_valid(grid, row, col, num):
    # Check row and column
    for i in range(9):
        if grid[row][i] == num or grid[i][col] == num:
            return False
    # Check subgrid
    subgrid_row, subgrid_col = 3 * (row // 3), 3 * (col // 3)
    for i in range(subgrid_row, subgrid_row + 3):
        for j in range(subgrid_col, subgrid_col + 3):
            if grid[i][j] == num:
                return False
    return True

# Backtracking solver
def solve_sudoku(grid):
    for row in range(9):
        for col in range(9):
            if grid[row][col] == 0:  # Empty cell
                for num in range(1, 10):
                    if is_valid(grid, row, col, num):
                        grid[row][col] = num
                        print(f"Placed {num} at ({row}, {col})")
                        print_grid(grid)
                        time.sleep(0.2)  # Pause for visualization

                        if solve_sudoku(grid):
                            return True
                        grid[row][col] = 0  # Backtrack
                        print(f"Backtracked at ({row}, {col})")
                        print_grid(grid)
                return False
    return True

# Function to generate a random Sudoku puzzle
def generate_sudoku():
    grid = [[0 for _ in range(9)] for _ in range(9)]
    for _ in range(random.randint(12, 18)):  # Fill a few random cells
        row, col = random.randint(0, 8), random.randint(0, 8)
        num = random.randint(1, 9)
        if is_valid(grid, row, col, num):
            grid[row][col] = num
    return grid

# Main program
random_sudoku = generate_sudoku()
print("Generated Sudoku puzzle:\n")
print_grid(random_sudoku)

if solve_sudoku(random_sudoku):
    print("Sudoku solved successfully!\n")
    print_grid(random_sudoku)
else:
    print("No solution exists for the given Sudoku.")
```

```python
import random
import networkx as nx
import matplotlib.pyplot as plt

# Function to format a grid for visualization in tree nodes
def format_grid(grid):
    return "\n".join(" ".join(str(num) if num != 0 else '.' for num in row) for row in grid)

# Function to print the Sudoku grid
def print_grid(grid):
    for row in grid:
        print(" ".join(str(num) if num != 0 else '.' for num in row))
    print("\n")

# Function to check if a number can be placed in a cell
def is_valid(grid, row, col, num):
    # Check row and column
    for i in range(9):
        if grid[row][i] == num or grid[i][col] == num:
            return False
    # Check subgrid
    subgrid_row, subgrid_col = 3 * (row // 3), 3 * (col // 3)
    for i in range(subgrid_row, subgrid_row + 3):
        for j in range(subgrid_col, subgrid_col + 3):
            if grid[i][j] == num:
                return False
    return True

# Backtracking solver with tree construction
def solve_sudoku(grid, tree, parent_node):
    for row in range(9):
        for col in range(9):
            if grid[row][col] == 0:  # Empty cell
                for num in range(1, 10):
                    if is_valid(grid, row, col, num):
                        grid[row][col] = num
                        current_state = format_grid(grid)
                        tree.add_node(current_state)
                        tree.add_edge(parent_node, current_state)

                        if solve_sudoku(grid, tree, current_state):
                            return True

                        grid[row][col] = 0  # Backtrack
                return False
    return True

# Function to generate a random Sudoku puzzle
def generate_sudoku():
    grid = [[0 for _ in range(9)] for _ in range(9)]
    for _ in range(random.randint(12, 18)):  # Fill a few random cells
        row, col = random.randint(0, 8), random.randint(0, 8)
        num = random.randint(1, 9)
        if is_valid(grid, row, col, num):
            grid[row][col] = num
    return grid



# Main program
random_sudoku = generate_sudoku()
print("Generated Sudoku puzzle:\n")
print_grid(random_sudoku)

# Create an empty tree for the state space
state_space_tree = nx.DiGraph()
initial_state = format_grid(random_sudoku)
state_space_tree.add_node(initial_state)

import time
start =time.time()
if solve_sudoku(random_sudoku, state_space_tree, initial_state):
    end = time.time()
    print(f"Sudoku solved successfully in {(end-start)//1000}s!\n")
    print_grid(random_sudoku)
    visualize_final_tree(state_space_tree)
else:
    print("No solution exists for the given Sudoku.")
```
